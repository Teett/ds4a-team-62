**Body Mass Index:**
BMI = Weight(kg) / [Size(m)] 2

```python
df['BMI'] = df['peso'] / (df['talla']*df['talla'])
df[['peso','talla','BMI']]
```

Creating Bins for BMI Distribution:

**Note:**
If your BMI is less than 18.5, it falls within the underweight range. If your BMI is 18.5 to 24.9, it falls within the normal or Healthy Weight range. If your BMI is 25.0 to 29.9, it falls within the overweight range. If your BMI is 30.0 or higher, it falls within the obese range.

```python
labels = ['Underweight','Normal or Helathy Weight','Overweight','Obese']
df['bin'] = pd.cut(df['BMI'], [18.5,24.9,25,29.9,100], labels = labels)
df[['BMI','bin']]
```

The great Ma

```python
axes = df['bin'].value_counts().plot.bar()
axes.set_title('BMI Classification Distribution')
axes.set_xlabel('BMI Classification');
```
